Generar los siguientes dataframes conteniendo:
    A.   Promedio por producto con el siguiente formato:           
    B. Mínimo, Máximo y Promedio de ventas por producto con el formato:     
    C.   Mínimo, Máximo y Promedio de ventas por producto con el formato: 
    D.    Mínimo, Máximo y Promedio de ventas por producto con el formato: 
    E.   Qué productos son provistos por los distintos fabricantes, indicando True/False con el siguiente formato:                  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import datetime
```

```python
tam=100
"""def crearLista(x):
    listaTipos=['transferencia','deposito','extraccion']
    listaReturn=[]
    for i in range(x):
        listaReturn.append(listaTipos[random.randint(0,2)])
    return listaReturn """

def random_date(start, end, tipo):
    """Generate a random datetime between `start` and `end`"""
    x = start + datetime.timedelta(seconds=random.randint(0, int((end - start).total_seconds())),
    ) 
    return x.date().month if tipo=='mes' else x.time()

l=[]
for x in range (tam): 
    l.append(random_date(pd.to_datetime('1/1/2008'),pd.to_datetime('1/1/2009'),'mes'))
"""h=[]
for x in range (tam): 
    h.append(random_date(pd.to_datetime('1/1/2008'),pd.to_datetime('1/1/2009'),'hora'))"""
    

alumn = {'CODIGOPRODUCTO':np.random.randint(1000,1050,size=tam),
         'CODIGOFABRICANTE':np.random.randint(200,250,size=tam),
        'MES':l,
        'VENTASMENSUALES':np.random.randint(0,100,size=tam)}
df = pd.DataFrame(alumn)
df.head()
```

A.   Promedio por producto con el siguiente formato:

```python
df.groupby('CODIGOPRODUCTO').agg({'VENTASMENSUALES':'sum'}).reset_index().rename(columns={'VENTASMENSUALES': 'Promedio de ventas'}).head()
```

B. Mínimo, Máximo y Promedio de ventas por producto con el formato:     

```python
df.groupby('CODIGOPRODUCTO').agg({'VENTASMENSUALES':['min','max','sum']}).stack()
```

 C. Mínimo, Máximo y Promedio de ventas por producto con el formato:

```python
df.groupby(['CODIGOPRODUCTO','MES']).agg({'VENTASMENSUALES':['min','max','mean']}).VENTASMENSUALES.\
rename(columns={'min':'Mínimo ventas','max':'Máximo ventas','mean':'Promedio ventas'}).reset_index().drop('MES', 1)
```

D.    Mínimo, Máximo y Promedio de ventas por producto con el formato: 

```python
df.groupby(['CODIGOPRODUCTO']).agg({'VENTASMENSUALES':['min','max','mean']}).VENTASMENSUALES.\
rename(columns={'CODIGOPRODUCTO':'Codigo producto','min':'Mínimo ventas','max':'Máximo ventas','mean':'Promedio ventas'}).reset_index().set_index('CODIGOPRODUCTO').T
```

E.   Qué productos son provistos por los distintos fabricantes, indicando True/False con el siguiente formato:

```python
nuevoDf = pd.DataFrame(columns=df['CODIGOFABRICANTE'].unique(),index=df['CODIGOPRODUCTO'].unique())
nuevoDf.head()
```

```python
grouped=df.groupby(['CODIGOFABRICANTE'])
```

```python
for columns in nuevoDf.columns:
    for index in nuevoDf.index:
        if index in grouped.get_group(columns).CODIGOPRODUCTO.values:
            nuevoDf.at[index,columns]=True
        else:
            nuevoDf.at[index,columns]=False
nuevoDf.head()
```
