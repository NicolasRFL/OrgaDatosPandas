```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import datetime
```

```python
tam=100
def crearLista(x):
    listaTipos=['Viajes de gulliver','Harry Potter AZK','Don Quijote','Calculo Tromba','Fuenteovejuna','Economia Mochon','AsnoDeOro',
               'Juguete Rabioso']
    listaReturn=[]
    for i in range(x):
        listaReturn.append(listaTipos[random.randint(0,len(listaTipos)-1)])
    return listaReturn 

l=[]
for x in range (tam): 
    l.append(random_date(pd.to_datetime('1/1/2008'),pd.to_datetime('1/1/2009'),'mes'))
"""h=[]
for x in range (tam): 
    h.append(random_date(pd.to_datetime('1/1/2008'),pd.to_datetime('1/1/2009'),'hora'))"""
    

ebooks = {'book_name':crearLista(tam),
         'book_id':np.random.randint(200,250,size=tam),
        'avg_rating':np.random.randint(0,4,size=tam)}
gr= pd.DataFrame(ebooks)
gr.head()
```

A. Obtener un DataFrame que indique el TOP5 de Ebooks en el sitio de Ebooks. (Para este punto se puede ignorar el segundo DataFrame). 

```python
top5eb=eb.groupby('book_id').agg({'rating':'mean'}).sort_values(by='rating',ascending=False).head(5).reset_index()
```

```python
top5eb
```

B. Obtener un DataFrame que indique qué libros tienen una diferencia de rating promedio mayor al 20% entre el sitio de Ebooks y GoodReads.

```python
mdf=eb.merge(gr,how='inner',on='book_id')
```

```python
mdf = mdf.groupby('book_id').agg({'rating':'mean','avg_rating':'mean'})
mdf
```

```python
mdf = ((mdf['rating']/mdf['avg_rating']>0.2)).to_frame()
```
